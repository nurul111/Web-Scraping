```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
```

```python
driver.get('https://www.amazon.com/')
```

```python
content = driver.page_source
#print(content)
soup = bs(content,'lxml')
#print(soup.prettify())
```

```python
All = soup.find_all('li',class_='feed-carousel-card')
print(All)
```

```python
BT=[] #List to store title of books
pri=[] #List to store price of books
pl=[] #List to store page length
Lan=[] #List to store language of the book
pub=[] #List to store publication date
Stock=[]# list to store Stock condition

for i in All:
    link=soup.find_all('span',class_='a-list-item')
    if link is None:
        break
    for l in link:
        if l is None:
            break
        href=l.find('a').get('href','')
        if href is None:
            break
        #href=l['href']
        #print(href)
        url='https://www.amazon.com/'+str(href)
        #print(url)
        driver.get(url)
        content = driver.page_source
        soup = bs(content,'lxml')
        #print(soup.prettify())
        BT1 = soup.find('span',class_='a-size-extra-large')
        if BT1 is None:
            break
        print(BT1.text.strip())
        pri1=soup.find('span',class_="a-size-base a-color-price a-color-price")
        print(pri1.text.strip())

        #pl1=soup.find_all('ol',class_="a-carousel")
        pl1=soup.find_all('div',class_="a-section a-spacing-none a-text-center rpi-attribute-value")
        bug_fix=soup.find_all('div',class_="a-section a-spacing-small a-text-center rpi-attribute-label")
        bug = []
        for i in bug_fix:
            bug.append(i.text.strip())
        #print(bug)
        new=[]
        for i in pl1:
            new.append(i.text.strip())
        #print(new)
        dic = dict()
        for i in range(len(bug)):
            dic[bug[i]]=new[i]
        print(dic)
        for key,value in dic.items():
            if key=='Print length' :
                        pl.append(value)
            elif key=='Language' :
                        Lan.append(value)
            elif key=='Publication date':
                        pub.append(value)
        
        BT.append(BT1.text.strip())
        #print(BT)
        pri.append(pri1.text.strip())
        #print(pri)
        
        
        st = soup.find('span',class_="a-size-medium a-color-success")
        #print(st.text.strip())
        if st is not None:
            Stock.append(st.text.strip())
        else: Stock.append('null')
        #print(Stock)
        new.clear()
    if BT1 is None:
        break

```

```python
df = pd.DataFrame({'Book title':BT,'Prices':pri,'Pages':pl,'language of the book':Lan,'Publication Date':pub,'Stock condition':Stock}) 
df.to_excel('top seller book of amazon.xlsx', index=False, encoding='utf-8')
df.head(20)
```
