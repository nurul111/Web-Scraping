```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
product_name=[]
product_price=[]
description=[]
review=[]
```

```python
for i in range(1,13):
    url = "https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3Drealme&p%5B%5D=facets.availability%255B%255D%3DExclude%2BOut%2Bof%2BStock&param=7564&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIlNob3AgTm93Il0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&otracker=clp_metro_expandable_1_3.metroExpandable.METRO_EXPANDABLE_Shop%2BNow_mobile-phones-store_Q1PDG4YW86MF_wp2&fm=neo%2Fmerchandising&iid=M_294f7644-82a0-4490-93c8-72767ef1405b_3.Q1PDG4YW86MF&ssid=96vz1pvquo0000001675864105889&sort=popularity&page="+str(i)
    r = requests.get(url)
    #print(r)
    soup = BeautifulSoup(r.text,'lxml')
    #print(soup.prettify())
    #np=soup.find('a',class_='_1LKTO3').get('href')
    #print(np)
    #cnp = "https://www.flipkart.com"+np
    #print(cnp)
    box=soup.find('div',class_='_1YokD2 _3Mn1Gg')
    names=box.find_all('div',class_='_4rR01T')
    #print(name)
    for i in names:
        name=i.text
        #print(name,'\n')
        product_name.append(name)
    #print(len(product_name))
    prices=box.find_all('div',class_='_30jeq3 _1_WHN1')
    for i in prices:
        price=i.text
        product_price.append(price)
    #print('\n',len(product_price))
    desc=box.find_all('ul',class_='_1xgFaf')
    for i in desc:
        desc=i.text
        description.append(desc)
    #print('\n',len(description))
    rev=box.find_all('div',class_='_3LWZlK')
    for i in rev:
        rev=i.text
        review.append(rev)
    #print('\n',len(review))
```

```python
df=pd.DataFrame({'Product Name':product_name,'Price':product_price,'Description':description,'Review':review})
df
```
