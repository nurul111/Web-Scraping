```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dve-shm-usage")
```

```python
driver = webdriver.Chrome(options = options)
```

# Course Link

```python
driver.get('https://outschool.com/')
time.sleep(3)
```

```python
url_list = []
```

```python
#classes by subjects
subjects = driver.find_elements(By.CSS_SELECTOR,'div.css-9pfejs div.css-0')[6]
```

```python
#sub_subject links
sub_subject = subjects.find_elements(By.CSS_SELECTOR,'div.css-1g360fb a')
count = len(sub_subject)

for x,sub in enumerate(sub_subject):
    if x==count-1:
        url_list.append(sub.get_attribute('href'))
#     if x==count-2:
#         break

print(url_list)


```

```python
df = pd.DataFrame({'Math and science_nature':url_list})
df
```

```python
df.to_csv('Math_Science_nature.csv')
```

# Teachers page

```python
def teachers(url):
    driver.get(url)
    time.sleep(3)
    
    #title
    try:
        Title = driver.find_element(By.CSS_SELECTOR,'
                                    ').text
    except:
        Title = 'Nan'
    
    #subtitle
    try:
        sub_title = driver.find_element(By.CSS_SELECTOR,'span.outschool-pjs51t').text
    except:
        sub_title = 'Nan'

    #ratings
    try:
        rating = driver.find_element(By.CSS_SELECTOR,'span.outschool-19brb1a').text
    except:
        rating = 'Nan'
   
    #review
    try:
        review = driver.find_element(By.CSS_SELECTOR,'span.outschool-19m73af').text
    except:
        review = 'Nan'
   
    #classes demand
    try:
        demand = driver.find_element(By.CSS_SELECTOR,'span.outschool-sywdfn').text
    except:
        demand = 'Nan'
    
    #classes taken
    try:
        classes = driver.find_elements(By.CSS_SELECTOR,'li.outschool-m1jbxv')[0].text
    except:
        classes = 'Nan'
   
    #by request
    try:
        req = driver.find_elements(By.CSS_SELECTOR,'li.outschool-m1jbxv')[3].text
    except:
        req = 'Nan'
   
    dictionary = {'Title':Title, 'Sub_Title':sub_title, 'Rating':rating, 'Review':review,
                  'Demand':demand, 'Classes_taken':classes, 'By_request':req, 'URL':url}
    return dictionary
    
    
```

```python
inp = input('Enter Teachers URL: ')
print(teachers(inp))
```

```python
df = pd.read_csv(r"C:\Users\Staff Asia\Downloads\Out_School_Math_Course_Info - Out_School_Course_Info.csv")
df
```

```python
df2 = pd.DataFrame({'Title':[], 'Sub_Title':[], 'Rating':[], 'Review':[],
                  'Demand':[], 'Classes_taken':[], 'By_request':[], 'URL':[]})
  
for i in df['Instructor_Link']:
    var = teachers(i)
    df2 = df2.append(var, ignore_index=True)
    
```

```python
df2.to_csv('Out_School_Math_Teachers_Info.csv', index = False)
```
